**Step 0:** Import the librarys and dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import ast
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
url_mov='https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv'
url_cred='https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv'

movies = pd.read_csv(url_mov)
credits = pd.read_csv(url_cred)
```

**Step 3:** Merge both dataset

```python
movies = movies.merge(credits, right_on=['movie_id','title'],left_on=['id','title']).drop('movie_id',axis=1)
```

**Step 4**: select the variables to be used.

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.sample(10)
```

**Step 5**: explore the final variables

```python
movies.info()
```

```python
movies.isna().sum()
```

'Overview' is the only variable that has missing values. Since it has only three NaN and I cannot guess or estimate what value it would have had, I will remove these three observations.

```python
movies.dropna(inplace=True)
movies.reset_index(inplace=True)
movies.isna().sum()
```

**Step 6**: transform the dataset from json to a tabular format

```python
# function to extraxt the name from any dict

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# function that extract the first three items from a dict
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
        count +=1  
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.sample(5)
```

```python
# function to extract only the Director's name

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.overview[0]
```

```python
#function to split the string by space with a comma
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.sample(5)
```

**Step 7**: remove the space from all the strings except title

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.sample(10)
```

**Step 8**: combine columns into one column named 'tags'

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

I create a new data frame with id, title and tags columns.

```python
new_df = movies[['id','title','tags']].copy()

new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df['tags'][0]
```

**Step 9**: tokenization of column 'tags'

```python
cv = CountVectorizer(max_features=5000 ,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[:10]
```

```python
cv.get_feature_names_out()
```

These are the words used on the tags for each movie.

**Step 10**: now we want to compare the different tags to find out those that are similar.

Cosine similarity allows you to measure the similarity of the documents. It will be a value between [0,1]. If it is 0 then both vectors are complete different. But in the place of that if it is 1, It will be completely similar.

Fuente: https://www.datasciencelearner.com/sklearn-cosine-similarity-implementation/

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity[0] # similarity between the label of the first film and the other films, one by one. 
```

```python
print('The six tags with the greatest similarity to the first are:')
sorted(list(enumerate(similarity[0])),reverse =True , key = lambda x:x[1])[1:6]
```

**Step 11**: create a recommendation function based on the cosine_similarity.

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] ##fetching the movie index
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate( distances)),reverse =True , key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
movie=new_df['title'][3]
print('The top six recommended movies for a person who likes', movie ,'are:')
recommend(movie)
```
