**Step 0:** Import the librarys and dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import ast
```

```python
url_mov='https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv'
url_cred='https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv'

movies = pd.read_csv(url_mov)
credits = pd.read_csv(url_cred)
```

It is observed that the dataset 'movies' has some missing data in some variables and that some of them have an 'object' data type when they should be categorical. Although, the dataset 'credits' appears to have no missing values, in the sample we see some empty lines (e.g., line 4247).

On the other hand, both tables seem to conform a documentary database, with some arrangements of embedded documents. The relationship between both collections seems to be the id in the movies collection and movie_id in the credits collection.

Next, both collections will be joined considering these ids together with title.

**Step 3:** Merge both dataset

```python
movies = movies.merge(credits, right_on=['movie_id','title'],left_on=['id','title']).drop('movie_id',axis=1)
```

**Step 4**: select the variables to be used.

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.sample(10)
```

**Step 5**: explore the final variables

```python
movies.info()
```

```python
movies.isna().sum()
```

'Overview' is the only variable that has missing values. Since it has only three NaN and I cannot guess or estimate what value it would have had, I will remove these three observations.

```python
movies.dropna(inplace=True)
movies.reset_index(inplace=True)
movies.isna().sum()
```

**Step 6**: transform the dataset from json to a tabular format

```python
# function to extraxt the name from any dict

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# function that extract the first three items from a dict
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
        count +=1  
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.sample(5)
```

```python
# function to extract only the Director's name

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.overview[0]
```

```python
#function to split the string by space with a comma
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.sample(5)
```

**Step 7**: remove the space from all the strings except title

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```
